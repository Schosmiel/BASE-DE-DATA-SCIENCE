## Faire une copie de df

```python
df_test = df.copy()
```

### Modifier la vaiable price_paid et Enlever le symbole $ par un caractére

```python
df_test.price_paid = df_test.price_paid.apply(lambda x : x.replace("$", ""))
```

###  Converti le prix en float

```python
df_test.price_paid = df_test.price_paid.astype(float)
```
