```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.describe().plot()
```

```python
df.price_paid = df.price_paid.apply(lambda x : x.replace("$", ""))
df.price_paid = df.price_paid.astype(float)
```

```python
df.dtypes
```

```python
df["price_paid"].mean()
```

```python
df["country"].value_counts().plot()
```

```python
df.groupby("gender").mean("price_paid").plot()
```

```python
df.groupby(["gender", "country"]).mean("price_paid").plot()
```

```python
df.groupby("date")["price_paid"].sum().plot(figsize=[10, 5])
```

```python
df.plot()
```

```python
df.groupby(["gender", "country"])["price_paid"].mean().plot.pie(figsize=[5, 5])
#df.groupby(["gender", "country"])["price_paid"].mean().plot.pie(legend=True)
```
