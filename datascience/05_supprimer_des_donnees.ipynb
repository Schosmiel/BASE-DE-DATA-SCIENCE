## Quelques notions de bases :

### axis=1 : Cela spécifie que les colonnes (axes horizontaux) doivent être traitées.

### axis=0 était utilisé, cela signifierait que les lignes (axes verticaux) seraient traitées.

### Supression de la colonne ip_address

```python
df.drop("ip_address", axis=1)
```

### La spécification de inplace veut dire que la modifiacation de s'exécuter directement sur le datafram(df)

```python
df.drop("ip_address", axis=1, inplace=True)
```
